```python
print('student_df') #concat
student_dict1 = {'Name': ['Joe','Nat'], 'Age': [20,21]}
student_df1 = pd.DataFrame(student_dict1)
print(student_df1)
```

```python
print('student_df')
student_dict2 = {'Name': ['Joe','Nat'], 'Age': [20,21]}
student_df2 = pd.DataFrame(student_dict2)
print(student_df2)
```

```python
print('combined with row')
combined_df = pd.concat([student_df1, student_df2], axis=0) #(axis=0 for row, axis=1 for column)
print(combined_df)
```

```python
print('combined with column')
combined_df = pd.concat([student_df1, student_df2], axis=1) #(axis=0 for row, axis=1 for column)
print(combined_df)
```

```python
print('student_df') #groupby
student_dict = {'Name': ['Joe','Nat','Harry'],
                'Class': ['A','B','A'],'Mark': [85.10,77.80,91.54]}
student_df = pd.DataFrame(student_dict)
print(student_df)
```

```python
print('Groupby sum')
student_df = student_df.groupby('Class').sum()
print(student_df)
```

```python
print('student_df') #aggregate
student_dict = {'Name': ['Joe','Nat','Harry'],
                'Class': ['A','B','A'],'Mark': [85.10,77.80,91.54]}
student_df = pd.DataFrame(student_dict)
print(student_df)
```

```python
print('aggregate')
student_stat = student_df['Mark']
print(student_stat)
student_stat = student_stat.agg(['sum','max','mean'])
print(student_stat)
```

```python
print('student_df') #transform
student_dict = {'Name': ['Joe','Nat','Harry'],
                'Class': ['A','B','A'],'Mark': [85.10,77.80,91.54]}
student_df = pd.DataFrame(student_dict)
print(student_df)

print('Transpose view')
print(student_df.T)
```
