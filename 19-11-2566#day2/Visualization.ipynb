```python
cols_2_remove = ['Product_Identifier','Supermarket_Identifier','Product_Supermarket_Identifier']
newdata = supermaket_data.drop(cols_2_remove, axis = 1)
```

```python
cat_cols = ['Product_Fat_Content','Product_Type','Supermarket _Size','Supermarket_Location_Type','Supermarket_Type']

num_cols = ['Product_Weight','Product_Shelf_Visibility','Product_Price','Supermarket_Opening_Year','Product_Supermarket_Sales']
```

```python
# Bar plot ดูความถี่
fig = plt.figure(figsize=(6,6))
ax = fig.gca()
counts = newdata['Product_Fat_Content'].value_counts()
counts.plot.bar(ax = ax)
ax.set_title('Bar plot for Product_Fat_Content')
```

```python
for col in cat_cols:
  fig = plt.figure(figsize=(6,6))
  ax = fig.gca()

  counts = newdata[col].value_counts()
  counts.plot.bar(ax=ax)
  ax.set_title('Bar plot for ' + col)
```

```python
# Scatter plot ดูการกระจายตัว
for col in num_cols:
  fig = plt.figure(figsize=(3,3))
  ax = fig.gca()

  newdata.plot.scatter(x=col, y='Product_Supermarket_Sales', ax=ax)
  #ax.set_title('Bar plot for ' + col)
```

```python
# box plot ดูการกระจายตัวของข้อมูล
for col in cat_cols:
  sns.boxplot(x=col, y = 'Product_Supermarket_Sales', data=newdata)
  plt.xlabel(col)
  plt.ylabel('Product Supermarket Sales')
  plt.show()
```

```python
# Heatmap
for col in cat_cols:
  corrmat = newdata.corr(numeric_only=True)

  f,ax = plt.subplots(figsize=(10,8))

  sns.heatmap(corrmat,square=True,annot=True,cmap='coolwarm',fmt='.2f')
  plt.show()
```
