Dataframe แบบข้อมูล Dictionary

```python
fruits_list = ['Apple',10,'Orange',55.50]
print(fruits_list)
```

```python
fruits_df = pd.DataFrame(fruits_list) #normal
fruits_df_index = pd.DataFrame(fruits_list, index=['Fruit1','Fruit2','Fruit3','Fruit4']) #index
fruits_df_index_columns = pd.DataFrame(fruits_list, index=['R0','R1','R2','R3'], columns=['Fruit']) #index+columns
print(fruits_df)
print(fruits_df_index)
print(fruits_df_index_columns)
```

```python
fruits_df_index.iloc[['0']]
```
