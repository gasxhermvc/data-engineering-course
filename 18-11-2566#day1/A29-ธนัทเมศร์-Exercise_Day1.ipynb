```python
import pandas as pd
```

```python
student_dict = { 'Name': ['Joe','Nat','Harry','Jack','Jose','Jill','Rose'],
                'Age': [20,21,19,17,18,19,17],
                 'Marks': [85.10,77.80,91.54,72,87.9,90,72]}
```

โจทย์1: prob 1 write a statement or statements to show the DataFrame df at row R1 and R5 in columns Name and Marks only.

```python
# create Dataframe from dict
df = pd.DataFrame(student_dict, index=['R1','R2','R3','R4','R5','R6','R7'])
print(df)
```

```python
# way 1
df.loc[['R1','R5'],['Name','Marks']]
```

```python
# way 2
df[['Name','Marks']].loc[['R1','R5']]
```

```python
# way 3
df[['Name','Marks']].iloc[[0,4]]
```
