# MySQL Connect

```python
import mysql.connector

conn = mysql.connector.connect(host="localhost",user="root",password="",database="lab2")
```

```python
if conn.is_connected():
    print("Connected...")
else:
    print("Lost connection...")
```

```python
def displayDataTables(cursor):
    # display all records
    table = cursor.fetchall()

    # describe table
    print('\n Table Description:')
    for attr in table:
	    print(attr)
```

```python
# create cursor instance
cursor = conn.cursor()
```

```python
# show databases
showDb = "show databases;"

cursor.execute(showDb)

displayDataTables(cursor)
```

```python
# create db
newDb = "create database testlab;"

cursor.execute(newDb)

displayDataTables(cursor)

# show databases
showDb = "show databases;"

cursor.execute(showDb)

displayDataTables(cursor)
```

```python
# drop db
newDb = "drop database testlab;"

cursor.execute(newDb)

displayDataTables(cursor)

# show databases
showDb = "show databases;"

cursor.execute(showDb)

displayDataTables(cursor)
```

```python
# show description
showDescr = "desc customer;"

cursor.execute(showDescr)

displayDataTables(cursor)
```

```python
# select customer all
customers = "select * from customer;"

cursor.execute(customers)

displayDataTables(cursor)
```

```python
# select customer id = 1
firstCustomer = "select * from customer where id = %s;"

cursor.execute(firstCustomer, ("1",))

displayDataTables(cursor)
```

```python
# insert new customer
newCustomer = "insert into customer (name,credit) values ( %s, %s)"
cursor.execute(newCustomer, ("Second two", "9999.99",))

displayDataTables(cursor)

# select customer all
customers = "select * from customer;"

cursor.execute(customers)

displayDataTables(cursor)
```

```python
# update customer
newCustomer = "update customer set name = %s where id = %s"
cursor.execute(newCustomer, ("Thanatmet T.", "0",))

displayDataTables(cursor)

# select customer all
customers = "select * from customer;"

cursor.execute(customers)

displayDataTables(cursor)
```

```python
# delete customer
deleteCustomer = "delete from customer where id = %s"
cursor.execute(deleteCustomer, ("0",))

displayDataTables(cursor)

# select customer all
customers = "select * from customer;"

cursor.execute(customers)

displayDataTables(cursor)
```

# Connect Mongodb

```python
import pymongo
client = pymongo.MongoClient('mongodb://localhost:27017/')
```

```python
db = client['customer']
```

```python
customer = db.employeeinformation
```

```python
record = {
    'name': 'First 1',
    'credit': 9999
}
customer.insert_one(record)
```

```python
records = [{
    'name': 'Two 2',
    'credit': 99
},
{
     'name': 'Three 3',
    'credit': 999
}]
```

```python
customer.insert_many(records)
```

```python
customer.find_one()
```

```python
for record in customer.find():
    print(record)
```

```python
for record in customer.find({'name':'One 1'}):
    print(record)
```

```python
for record in customer.find({'name':{'$in':['One 1','Two 2']}}):
    print(record)
```

```python
for record in customer.find({'credit':{'$lt':999}}):
    print(record)
```
